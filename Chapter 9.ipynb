```python
import pandas as pd
from datetime import datetime
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('examples/spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']

spx.plot(ax=ax, style='k-')

crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor='black', headwidth=4, width=2, headlength=4),
                horizontalalignment='left', verticalalignment='top')

# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])

ax.set_title('Important dates in the 2008-2009 financial crisis')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color='g', alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

```python
tips = pd.read_csv('examples/tips.csv')
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts
```

```python
# Not many 1- and 6- person parties
party_counts = party_counts.loc[:, 2:5]
```

```python
# Normalize to sum to 1
party_pcts = party_counts.div(party_counts.sum(1), axis=0)
party_pcts
```

```python
# Fraction of parties by size on each day
party_pcts.plot.bar()
```

```python
import seaborn as sns
fig = plt.figure()
# Tipping percentage by day with error bars
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips.head()
```

```python
sns.barplot(x='tip_pct', y='day', data=tips, orient='h')
```

```python
fig = plt.figure()
# Tipping percentage by day and time
sns.barplot(x='tip_pct', y='day', hue='time', data=tips, orient='h')
```

```python
fig = plt.figure()
# Histogram of tip percentages
tips['tip_pct'].plot.hist(bins=50)
```

```python
fig = plt.figure()
# Density plot of tip percentages
tips['tip_pct'].plot.density()
```

```python
fig = plt.figure()
# Normalized histogram of normal mixture with density estimate
comp1 = np.random.normal(0, 1, size=200)
comp2 = np.random.normal(10, 2, size=200)
values = pd.Series(np.concatenate([comp1, comp2]))
sns.distplot(values, bins=100, color='k')
```

```python
macro = pd.read_csv('examples/macrodata.csv')

data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna()
trans_data[-5:]
```

```python
fig = plt.figure()
# A seaborn regression/scatter plot
sns.regplot('m1', 'unemp', data=trans_data)
plt.title('Changes in log %s vs log %s' % ('m1', 'unemp'))
```

```python
# Pair plot matrix of statsmodels macro data
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.2})
```

```python
# Tipping percentage by day/time/smoker
sns.factorplot(x='day', y='tip_pct', hue='time', col='smoker', kind='bar', data=tips[tips.tip_pct < 1])
```

```python
# tip_pct by day; facet by time/smoker
sns.factorplot(x='day', y='tip_pct', row='time', col='smoker', kind='bar', data=tips[tips.tip_pct < 1])
```
